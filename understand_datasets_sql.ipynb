```python
#list community area names and numbers with per capita income less than 11000
%sql select COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME, PER_CAPITA_INCOME from Chicago_Census_Data where PER_CAPITA_INCOME < 11000 
```

```python
#list of columns inthe `Chicago Public Schools` table
%sql select name, type, length(type) from PRAGMA_TABLE_INFO('Chicago_Public_Schools')
```

```python
#list all cases for crimes involving minors
%sql select CASE_NUMBER, COMMUNITY_AREA_NUMBER, DESCRIPTION from Chicago_Crime_Data where DESCRIPTION LIKE '%minor%'
```

```python
#list of all kidnapping crimes involving a child
%sql select CASE_NUMBER, DESCRIPTION from Chicago_Crime_Data where DESCRIPTION like '%child%'
```

```python
#number of crimes recorded at schools
%sql select DISTINCT COUNT(CASE_NUMBER) from Chicago_Crime_Data where LOCATION_DESCRIPTION like '%school%'
```

```python
#identifying cases all kidnapping cases involving a child
%sql select * from Chicago_Crime_Data where PRIMARY_TYPE = 'KIDNAPPING' and DESCRIPTION LIKE '%CHILD%'
```

```python
#identifying cases involving a minor
%sql select * from Chicago_Crime_Data where PRIMARY_TYPE = 'KIDNAPPING' and DESCRIPTION LIKE '%MINOR%'
```

```python
#list of 5 community areas with the highest % of households below poverty line
%sql select COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY from Chicago_Census_Data order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc LIMIT 5
```

```python
%sql select COMMUNITY_AREA_NUMBER, SAFETY_SCORE from Chicago_Public_Schools where (select MIN(SAFETY_SCORE) from Chicago_Public_Schools) order by SAFETY_SCORE desc limit 10 
```

```python
#avg safety score from middle schools
%sql select AVG(SAFETY_SCORE) from Chicago_Public_Schools where "Elementary, Middle, or High School" = "MS"
```

```python
#community area with most criminal incidents
%sql select DISTINCT COMMUNITY_AREA_NUMBER, COUNT(CASE_NUMBER) from Chicago_Crime_Data where (select MIN(SAFETY_SCORE) from Chicago_Public_Schools) group by COMMUNITY_AREA_NUMBER order by COUNT(CASE_NUMBER) desc
```

```python
#list of kind of crimes recorded in schools
%sql select DISTINCT CASE_NUMBER, PRIMARY_TYPE, LOCATION_DESCRIPTION, DESCRIPTION from Chicago_Crime_Data where LOCATION_DESCRIPTION like '%school%'
```

```python
#List of type of schools along with the average safety score for each type
%sql select "Elementary, Middle, or High School", AVG(SAFETY_SCORE) as AVG_SAFETY_SCORE from Chicago_Public_Schools group by "Elementary, Middle, or High School"
```

```python
#5 community areas with the highest % of households below poverty line
%sql select COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY from Chicago_Census_Data order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc limit 5
```

```python
#community area most crime prone
%sql select COMMUNITY_AREA_NUMBER, COUNT(COMMUNITY_AREA_NUMBER) as MOST_CRIME_PRONE from Chicago_Crime_Data group by COMMUNITY_AREA_NUMBER order by COUNT(COMMUNITY_AREA_NUMBER) desc limit 1
```

```python
#subquery to find the name of the community are with highest hardship index
%sql select COMMUNITY_AREA_NAME, HARDSHIP_INDEX from Chicago_Census_Data where ( select MAX(HARDSHIP_INDEX) from Chicago_Census_Data) order by HARDSHIP_INDEX desc limit 5
```

```python
#subquery to determine the community area name with most number of crimes
%sql select COMMUNITY_AREA_NAME from Chicago_Census_Data where COMMUNITY_AREA_NUMBER = (select COMMUNITY_AREA_NUMBER from Chicago_Crime_Data group by COMMUNITY_AREA_NUMBER order by COUNT(COMMUNITY_AREA_NUMBER) desc limit 5) limit 5
```

```python
#subquery to find the name of community are with the highest hardship index
%sql select COMMUNITY_AREA_NAME from Chicago_Census_Data where HARDSHIP_INDEX in (select MAX(HARDSHIP_INDEX) from Chicago_Census_Data)
```
